## Important Patterns 

```python
t = 'The catfish is an animal. The catdog is not.'
```

```python
re.search('cat', t)
```

```python
re.findall('cat', t)
```

```python
re.findall('cat(?!dog)', t)  # lookahead assertion
```

```python
u = '1101 is binary, but not 8765'
```

```python
re.findall('\d+', u)
```

```python
re.findall('(?<=binary, but not )\d+', u)  # lookbehind assertion
```

```python
re.findall('\w', u)[:6]
```

```python
re.findall('\w+', u)
```

```python
re.split('\s', u)
```

```python
v = 'Our email address is training@tpq.io. And you can use team@tpq.io. notemail@xxx'
```

```python
re.search('\w+@', v)
```

```python
re.search('(\w+)@', v)
```

```python
re.findall('\w+@', v)
```

```python
re.findall('(\w+)@', v)
```

```python
re.findall('(\w+)@(\w+)', v)
```

```python
re.findall('(\w+)@(\w+)\.(\w+)', v)
```

```python
re.findall('(\w+)@(\w+\.\w+)', v)
```

```python
matches = re.findall('(\w+)@(\w+\.\w+)', v)
```

```python
matches
```

```python
[m[0] for m in matches]
```

```python
re.findall('(?P<users>\w+)@(?P<domains>\w+\.\w+)', v)
```

```python
match = re.search('(?P<users>\w+)@(?P<domains>\w+\.\w+)', v)
```

```python
match.groups()
```

```python
match.group('users')
```

```python
match.group('domains')
```
