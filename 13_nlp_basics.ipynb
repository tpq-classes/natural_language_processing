## Predicting the Alphabet

```python
ab = ab[:-2]
```

```python
abi = np.array([cti[c] for c in ab])
abi[:10]
```

```python
lags = 3
```

```python
def create_matrix(n, lags=lags):
    data = list()
    for i in range(0, len(n) - lags):
        data.append(n[i:i+lags+1])
    return np.array(data)
```

```python
data = create_matrix(abi)
data[:3]
```

```python
def generate_model(hu=24, lags=lags):
    model = Sequential()
    model.add(Dense(hu, activation='relu', input_dim=lags))
    model.add(Dense(hu, activation='relu'))
    model.add(Dense(1, activation='linear'))
    model.compile(loss='mse', optimizer=Adam())
    return model
```

```python
model = generate_model()
```

```python
f = data[:, :lags]
l = data[:, lags]
```

```python
%time model.fit(f, l, epochs=250, verbose=False)
```

```python
model.predict(f)[:5].flatten()
```

```python
model.predict(f)[:5].round().flatten()
```

```python
model.evaluate(f, l)
```

```python
sample = 'lmn'
```

```python
sample_ = np.array([cti[c] for c in sample])
sample_
```

```python
p = model.predict(sample_.reshape(1, -1))
p
```

```python
p[0, 0]
```

```python
itc[round(p[0, 0])]
```

## Predicting Characters in Words

```python
import requests
```

```python
text = requests.get('https://hilpisch.com/walden.txt').text
```

```python
text[:500]
```

```python
text = text.replace('\n', ' ')
```

```python
tokens = text.split()
tokens[:5]
```

```python
tokens = [t.lower() for t in tokens if len(t) > 3]
```

```python
tokens[:5]
```

```python
words = [t for t in tokens if len(t) == 4 and t.isalpha()]
```

```python
len(words)
```

```python
words[:10]
```

```python
len(set(words))
```

```python
words = sorted(list(set(words)))
```

```python
words[50:60]
```

```python
wl = list()
for w in words:
    il = list()
    for c in w:
        il.append(cti[c])
    wl.append(il)
wl = np.array(wl)
```

```python
wl[50:60]
```

```python
n = 500
```

```python
words[n:n + 10]
```

```python
data = wl[n:n + 10]
data
```

```python
f = data[:, :lags]
l = data[:, lags]
```

```python
model = generate_model(hu=128)
```

```python
%time model.fit(f, l, epochs=2500, verbose=False)
```

```python
model.evaluate(f, l)
```

```python
sample = 'nic'
# sample = 'nin'
# sample = 'new'
```

```python
sample_ = np.array([cti[c] for c in sample])
sample_
```

```python
p = model.predict(sample_.reshape(1, -1))
p
```

```python
p[0, 0]
```

```python
itc[round(p[0, 0])]
```
