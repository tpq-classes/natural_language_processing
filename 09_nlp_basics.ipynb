## Non-Greedy Matching

```python
re.findall('\w*c.*t\w*', s, re.IGNORECASE)
```

```python
html = '<tag>This might be <b>some</b> web text.</tag>'
```

```python
re.search('<.+>', html)  # greedy matching
```

```python
re.search('<.+?>', html)  # non-greedy matching
```

```python
re.findall('<.+?>', html)  # non-greedy matching
```

```python
re.sub('<.+?>', '', html)  # non-greedy matching
```

```python
re.sub('<.+>', '', html)  # greedy matching
```

```python
re.findall('<(.+?)>', html)  # non-greedy matching
```

## Use Cases

```python
with open('/content/natural_language_processing/email_addresses.txt', 'r') as f:
    ea = f.read()
```

```python
# print(ea)
```

```python
ep = re.compile('\w+@\w+\.\w+')
```

```python
len(ea.split('\n'))
```

```python
len(ep.findall(ea))
```

```python
ea.split('\n')[-5:]
```

```python
ep.findall(ea)[-5:]
```

```python
ep = re.compile('[a-zA-Z0-9.+_\-]+@[a-zA-Z0-9.\-]+\.\w+')
```

```python
matches = ep.findall(ea)
matches
```

```python
len(matches)
```

```python
ea.split('\n') == matches
```
