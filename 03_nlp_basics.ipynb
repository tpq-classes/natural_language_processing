## Large Data File (CSV)

```python
url = 'https://certificate.tpq.io/findata.csv'
```

```python
raw = requests.get(url).text
```

```python
len(raw)
```

```python
raw[:500]
```

```python
print(raw[:500])
```

```python
rows = raw.split('\n')
```

```python
rows[:7]
```

```python
len(rows)
```

```python
values = [row.split(',') for row in rows]
```

```python
values[:3]
```

```python
values[-3:]
```

```python
values[0]
```

```python
EUR = [float(row[9]) for row in values[1:-1]]
```

```python
EUR[:10]
```

```python
sum(EUR) / len(EUR)
```

```python
dates = [row[0] for row in values[1:-1]]
```

```python
dates[:5]
```

```python
import pandas as pd
```

```python
dates_ = [pd.Timestamp(d) for d in dates]
```

```python
from pylab import plt
plt.style.use('seaborn-v0_8')
```

```python
plt.plot(dates_, EUR);
```
